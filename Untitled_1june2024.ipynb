```python
# def get_ce_pe_price(row):
    
#     if 'CE' in row['CE']:
# #         print('yes')
#         ce_ticker = row['CE']
#     else:
#         ce_ticker = row['PE']
# #         print('NO')
#     time = row['time']
#     price = df.loc[(df['Ticker'] == ce_ticker) & (df['Time'] == time), 'Close'].values
# #     return price.values[0] if not price.empty else np.nan
#     print(price[0])
# #     return price

# df_fut.apply(get_ce_pe_price, axis=1)
# # print(pri)
# # df_fut['pe_price'] = df_fut.apply(get_ce_pe_price, axis=1)
# # df_fut.to_csv('try8.csv')

def ceiling_xcl(x, y):
    if y == 0:
        raise ValueError("The divisor 'y' cannot be zero.")
    remainder = x % y
    if remainder == 0:
        return int(x)
    else:
        return int(x + (y - remainder))

def get_ce_pe_price(row, type):
    # ce_ticker = row['CE']
    if type == 'CE':
        ce_ticker = row['CE']
    if type == 'PE':
        ce_ticker = row['PE']
    time = row['time']
    price = df.loc[(df['Ticker'] == ce_ticker) & (df['Time'] == time), 'Close'].values[0]
    return price

def get_greeks(row, strike, opt_type, int_rate = 0):
    underlying_price = row['forward']
    strike = strike
    exp = row['TTE']
    if opt_type == 'CE':
        call_price = row['ce_price']
        try:
            calc = mibian.BS([underlying_price, strike, int_rate, exp], callPrice = call_price)
            IV = calc.impliedVolatility
            c = mibian.BS([underlying_price, strike, int_rate, exp], volatility = IV)
    #         return pd.concat([new_col], row)
            print(f'\n call {underlying_price}, {strike}, {exp}, {call_price}, {IV}')
            return IV, c.callDelta, c.gamma, c.callTheta, c.vega
        except Exception as e:
            print(f"Error calculating greeks for CE: {e}")
            return np.nan, np.nan, np.nan, np.nan
    if opt_type == 'PE':
        put_price = row['pe_price']
        try:
            calc = mibian.BS([underlying_price, strike, int_rate, exp], putPrice = put_price)
            IV = calc.impliedVolatility
            c = mibian.BS([underlying_price, strike, int_rate, exp], volatility = IV)
            print(f'\n put {underlying_price}, {strike}, {exp}, {put_price}, {IV}')
            return IV, c.putDelta, c.gamma, c.putTheta, c.vega
        except Exception as e:
            print(f"Error calculating greeks for PE: {e}")
            return np.nan, np.nan, np.nan, np.nan
```

```python
df_fut['straddle'] = -50000
df_fut['delta'] = (df_fut['straddle'] * df_fut['ce_delta']) + (df_fut['straddle'] * df_fut['pe_delta'])
df_fut['gamma'] = (df_fut['straddle'] * df_fut['ce_gamma']) + (df_fut['straddle'] * df_fut['pe_gamma'])
df_fut['theta'] = (df_fut['straddle'] * df_fut['ce_theta']) + (df_fut['straddle'] * df_fut['pe_theta'])
df_fut['vega'] = (df_fut['straddle'] * df_fut['ce_vega']) + (df_fut['straddle'] * df_fut['pe_vega'])
```

# getting hedge point and D/G

```python
# hedge = sqrt(straddle_theta / 2 * abs(straddle_gamma))
df_fut['hedge_point'] = np.sqrt(df_fut['theta'] / (2 * abs(df_fut['gamma'])))
df_fut['D/G'] = df_fut['delta'] / df_fut['gamma']
```

# finding the row index where the check condition is met

```python
# condition = abs(d/g) > hedge, chk else ''
df_fut['check'] = np.nan
condition = abs(df_fut['D/G']) > df_fut.loc[row_index,'hedge_point']
df_fut.loc[condition, 'check'] = 'CHK'
chk_row_index = df_fut.index[condition].tolist()
print(f'\n index where condition is met - {chk_row_index}')
```
