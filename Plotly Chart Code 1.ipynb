```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df['time'] is in string format, convert it to datetime
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace = True)
time_col = []
for i in range(len(df)):
    time_col.append(df.index[i].time())
df['time'] = time_col



# Plot Forward and Use Strike
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('Forward / Use Strike', color=color)
ax1.plot(df['time'], df['forward'], color=color, label='Forward')
ax1.scatter(df['time'], df['use_strike'], color=color, label='Use Strike', marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# Plot Final PnL
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Final PnL', color=color)
ax2.plot(df['time'], df['final_pnl'], color=color, label='Final PnL')
ax2.tick_params(axis='y', labelcolor=color)

# Filter timestamps to show only one timestamp per hour
hourly_timestamps = df['time'].dt.floor('H').unique()

# Set x-axis ticks at hourly intervals
ax1.set_xticks(hourly_timestamps)
ax1.set_xticklabels(hourly_timestamps.strftime('%H:%M:%S'), rotation=45)

# Format x-axis
plt.title('Forward, Final PnL, and Use Strike over Time')

# Add legend
fig.tight_layout()
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=3)

plt.show()
```
